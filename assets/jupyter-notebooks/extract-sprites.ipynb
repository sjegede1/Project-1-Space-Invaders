### Function to create action images

```python
def extract_sprites(character):
    name = character['name']
    actions = character['actions']
#     character['map'] = pd.read_csv(f'./sprites/{name}/{name}-map.csv')
    for action in actions:
        action_num = 1
        for sprite_idx in character[action]:
            _,left,upper,width,height = character['map'].iloc[sprite_idx-1]
            right=left+width
            lower=upper+height
            im = Image.open(f"C:/Users/sjege/Documents/GitHub/Project1/assets/sprites/sprite-sheets/{name}-transparent.png").crop((left, upper, right, lower))
            im.save(f"C:/Users/sjege/Documents/GitHub/Project1/assets/sprites/{name}/{name}-{action}-{action_num}.png")
            action_num+=1
            
            
```

```python
for character in characters:
    extract_sprites(character)
```
